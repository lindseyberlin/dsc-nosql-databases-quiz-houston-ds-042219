## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

A SQL database is a relational database, which stores data that you know the basic form of (in terms of what categories the data belongs in, and how wide (how many columns) the data will be (have)) before you begin to store the data. NoSQL databases have varied structures, potentially disassembling the data and tasks in order to make those tasks run more smoothly.





## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

A NoSQL database may be the right choice for your data if you are not sure how many categories it may have, or if you need to store many different types of data of wildly varying forms.

## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance is the practice of creating redundancy in order to prevent distruptions in the event of a server going down, and RDDs are datasets that are fault tolerant - they have pieces of the dataset distributed across multiple servers, with redundant backups available in case a server or two go down.
